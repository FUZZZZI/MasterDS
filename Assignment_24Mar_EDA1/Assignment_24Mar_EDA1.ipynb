### Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

```python
wine.isnull().sum()
```

Currently, there is no missing in the wine dataset.

If there is missing value, I can use imputing method like:

1. mean, median to fill the **continuous feature** and mode to fill the **discrete features**.
2. Imputer algorithms from sklearn library like SimpleImputer, KNNImputer.

Types of imputation and their advantages & disadvantages:<br>
**Mean imputation:** This involves replacing missing values with the mean of the non-missing values. One advantage of this method is that it is simple and easy to implement. However, a major disadvantage is that it can introduce bias, especially if the missing data is not missing completely at random. Additionally, mean imputation may not be appropriate for variables with extreme values or for data that is not normally distributed.

**Hot deck imputation:** This involves filling in missing data with a value from another similar record in the same dataset. The advantage of this method is that it preserves the relationship between variables and can be effective when the missing data is related to other variables in the dataset. However, it can be computationally expensive and requires a sufficient number of similar records to work effectively.

**Multiple imputation:** This involves generating multiple plausible values for missing data and using them to create multiple completed datasets. The advantage of this method is that it accounts for the uncertainty of the missing data and produces more accurate estimates of standard errors. However, it can be computationally intensive and may not be appropriate for datasets with a large proportion of missing data.

**Regression imputation:** This involves using regression models to estimate the missing values based on the relationships between the missing data and other variables in the dataset. The advantage of this method is that it can capture the complex relationships between variables and produce accurate estimates of missing data. However, it requires the variables used in the regression to be correlated with the missing data, and it may not be appropriate for non-linear relationships.

**K-nearest neighbor imputation:** This involves identifying the k most similar records to the record with missing data and using their values to impute the missing value. The advantage of this method is that it can preserve the correlation structure of the data and can be effective when the missing data is related to other variables in the dataset. However, it can be computationally expensive and requires a sufficient number of similar records to work effectively.

### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
wine_data_std = scaler.fit_transform(wine)
```

```python
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(wine_data_std)
```

We can examine the explained variance ratio of each principal component using the explained_variance_ratio_ attribute of the PCA object:

```python
explained_variance_per=pca.explained_variance_ratio_*100
explained_variance_per
```

From the explained variance ratios, we can see that the first principal component explains the highest amount of variance (26.0%), followed by the second and third principal components (18.6% and 14.0%, respectively).

To determine the minimum number of principal components required to explain 90% of the variance in the data, we can calculate the cumulative explained variance ratio and find the index of the first principal component that exceeds 90%.

```python
cumulative=0
n=0
for i in range(len(explained_variance_per)):
    if cumulative>90:
        break
    else:
        cumulative += explained_variance_per[i]
    n +=1
print("Minimum number of principal components required to explain 90% of variance: ",n)
```
